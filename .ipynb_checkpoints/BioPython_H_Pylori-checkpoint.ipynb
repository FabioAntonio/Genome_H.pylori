Importação das Bibliotecas

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import matplotlib.pyplot as plt
from collections import Counter
```

Abertura e leitura do arquivo

```python
arquivo = 'GCF_017821535.1_ASM1782153v1_genomic.fna'

try:
    with open(arquivo, 'r') as handle:
        registros = list(SeqIO.parse(handle, 'fasta'))  # Use SeqIO.parse para múltiplos registros
        if not registros:
            raise ValueError("No records found in handle")
        for registro in registros:
            print(f"ID: {registro.id}")
            print(f"Sequência: {registro.seq}")
except ValueError as e:
    print(f"Erro: {e}")
except FileNotFoundError:
    print("Arquivo não encontrado. Verifique o caminho do arquivo.")
except Exception as e:
    print(f"Um erro ocorreu: {e}")
```

# Gerando gráficos

Analisando a frequência de nucleotídeos ao longo da sequência

```python
# Função para calcular a frequência acumulada de um nucleotídeo ao longo da sequência
def calculate_nucleotide_distribution(sequence, nucleotide):
    return [sequence[:i+1].count(nucleotide) for i in range(len(sequence))]

# Sequência de nucleotídeos de exemplo
sequence = "ATCGATTGAGCTCTAGCG"

# Calculando a distribuição para cada nucleotídeo
distribution_A = calculate_nucleotide_distribution(sequence, 'A')
distribution_T = calculate_nucleotide_distribution(sequence, 'T')
distribution_C = calculate_nucleotide_distribution(sequence, 'C')
distribution_G = calculate_nucleotide_distribution(sequence, 'G')

# Criando o gráfico de linhas
plt.plot(distribution_A, label='A', color='blue')
plt.plot(distribution_T, label='T', color='orange')
plt.plot(distribution_C, label='C', color='green')
plt.plot(distribution_G, label='G', color='red')
plt.xlabel('Posição na Sequência')
plt.ylabel('Frequência Acumulada')
plt.title('Distribuição de Nucleotídeos ao Longo da Sequência')
plt.legend()
plt.show()
```

Analisando a  frequência de subsequências de comprimento k (k-mers)

```python
# Função para gerar k-mers
def generate_kmers(seq_record, k):
    kmers = [sequence[i:i+k] for i in range(len(sequence)-k+1)]
    return kmers

# Gerando k-mers de tamanho 2
k = 2
kmers = generate_kmers(sequence, k)

# Contando a frequência dos k-mers
kmer_count = Counter(kmers)

# Preparando os dados para o gráfico
labels = kmer_count.keys()
counts = kmer_count.values()

# Criando o gráfico de barras
plt.bar(labels, counts, color='purple')
plt.xlabel('K-mer')
plt.ylabel('Frequência')
plt.title(f'Frequência de {k}-mers na Sequência')
plt.show()
```
